```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'vertical_avg_mbe': [-0.392, -0.108, -0.08, 0.290, -0.092], 'vertical_avg_rmse': [1.829, 0.775, 0.883, 0.624, 1.011], 'surface_mbe': [-0.68, -0.178, 0.198, 0.291, 0.3245], 'surface_rmse': [1.708, 0.752, 0.747, 0.610, 0.909], 'toa_mbe': [-0.07, -0.121, -0.118, 0.3135, -0.074], 'toa_rmse': [1.22, 0.718, 0.617, 0.621, 0.473]})
```

```python
df.index = ['CNN', 'MLP', 'GCN', 'GN', 'LGCN'
]
```

```python
df['cpu_time'] = [3.0288, 0.2277, 4.297, 6.969, 4.297]
```

```python
df['gpu_time'] = [0.3894, 0.0066, 2.879, 0.6850, 2.879]
```

```python
df
```

```python
df.plot.line(x='cpu_time', y='vertical_avg_mbe')
```

```python
df.plot.line(x='cpu_time', y='vertical_avg_rmse')
```

```python
df.plot.line(x='gpu_time', y='vertical_avg_mbe')
```

```python
df.plot.line(x='cpu_time', y='vertical_avg_rmse')
```

```python
df['reported_vertical_avg_mbe'] = [-0.142, -0.160, 0.034, -0.001, -0.179]
df['reported_vertical_avg_rmse'] = [0.303, 0.701, 1.209, 0.648, 0.878]
df['reported_surface_rmse'] = [0.265, 0.684, 1.26, 0.620, 0.714]
df['reported_surface_mbe'] = [-0.138, -0.282, 0.244, 0.017, -0.241]
df['reported_toa_mbe'] = [-0.177, -0.208, 0.071, -0.033, -0.048]
df['reported_toa_rmse'] = [0.284, 0.573, 0.815, 0.434, 0.440]
```

```python
df['reported_cpu_time'] = [3.187, 0.164, 4.684, 28.12, 4.684]
df['reported_gpu_time'] = [0.0218, 0.0016, 1.6818, 0.1659, 1.6818]
```

```python
df
```

```python
ax1 = df.plot(kind='scatter', x='cpu_time', y='vertical_avg_rmse', color='r', figsize=(10,10), label='ours')    
ax2 = df.plot(kind='scatter', x='reported_cpu_time', y='reported_vertical_avg_rmse', color='g', ax=ax1, label='reported')    
for idx, row in df.iterrows():
    ax1.annotate(idx, (row['cpu_time'], row['vertical_avg_rmse']), fontsize=18)
    ax1.annotate(idx, (row['reported_cpu_time'], row['reported_vertical_avg_rmse']), fontsize=18 )
ax1.set_xlabel('CPU runtime', fontsize=20)
ax1.xaxis.set_tick_params(labelsize=16)
ax1.set_ylabel('Avg RMSE', fontsize=20)
ax1.yaxis.set_tick_params(labelsize=16)
ax1.legend(fontsize=20)
```

```python
ax1 = df.plot(kind='scatter', x='gpu_time', y='vertical_avg_rmse', color='r', figsize=(10,10), label='ours')    
ax2 = df.plot(kind='scatter', x='reported_gpu_time', y='reported_vertical_avg_rmse', color='g', ax=ax1, label='reported')    
for idx, row in df.iterrows():
    ax1.annotate(idx, (row['gpu_time'], row['vertical_avg_rmse']), fontsize=18)
    ax1.annotate(idx, (row['reported_gpu_time'], row['reported_vertical_avg_rmse']), fontsize=18 )
ax1.set_xlabel('GPU runtime', fontsize=20)
ax1.xaxis.set_tick_params(labelsize=16)
ax1.set_ylabel('Avg RMSE', fontsize=20)
ax1.yaxis.set_tick_params(labelsize=16)
ax1.legend(fontsize=20)
```

```python
ax1 = df.plot(kind='scatter', x='gpu_time', y='vertical_avg_mbe', color='r', figsize=(10,10), label='ours')    
ax2 = df.plot(kind='scatter', x='reported_gpu_time', y='reported_vertical_avg_mbe', color='g', ax=ax1, label='reported')    
for idx, row in df.iterrows():
    ax1.annotate(idx, (row['gpu_time'], row['vertical_avg_mbe']), fontsize=18)
    ax1.annotate(idx, (row['reported_gpu_time'], row['reported_vertical_avg_mbe']), fontsize=18 )
ax1.set_xlabel('GPU runtime', fontsize=20)
ax1.xaxis.set_tick_params(labelsize=16)
ax1.set_ylabel('Avg MBE', fontsize=20)
ax1.yaxis.set_tick_params(labelsize=16)
ax1.legend(fontsize=20)
```
